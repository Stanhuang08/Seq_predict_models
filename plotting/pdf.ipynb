```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.figure_factory as ff
```

```python
minmax_mse = np.load('minmax_mse.npy')
standard_mse = np.load('standard_mse.npy')
```

```python
hist_data = [standard_mse, minmax_mse]

group_labels = ['standard_scaler', 'minmax_scaler']
colors = ['#37AA9C', '#F75000'] #  '#94F3E4'

# Create distplot with curve_type set to 'normal'
fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)

# Add title
fig.update_layout(
    title_text='PDF of MSE',
    xaxis_title='MSE',
    yaxis_title='Density',
    xaxis_range=[1, 100]
)
fig.show()
```
