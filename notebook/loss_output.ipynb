```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_loss = np.load('train_loss_epochs.npy')
vali_loss = np.load('vali_loss_epochs.npy')
test_loss = np.load('test_loss_epochs.npy')
```

```python
epoch = len(vali_loss)
x = range(int(epoch))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, train_loss, label='train loss')
plt.plot(x, vali_loss, label='vali loss')
plt.plot(x, test_loss, label='test loss')
plt.legend()
plt.xlabel('epochs', fontsize=12)
plt.ylabel('MSE', fontsize=12)
plt.legend(fontsize=10)
plt.savefig('loss.png') 
plt.show()
```
